```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv('GroceryStoreDataSet.csv')
df.head()
```

```python
#Since it's missing a header we'll create one, the new column header 'Products' 
df=pd.read_csv('GroceryStoreDataSet.csv', names=['products'])
df.head() 
```

```python
df.shape
```

```python
#following code converts a pandas dataframe into a list of lists.
data=list(df["products"].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True to 0 & 1 for easy readability
df=df.replace(False,0)
df=df.replace(True,1)

df
                
```

```python
df=apriori(df,min_support=0.2,use_colnames=True)
df
```

```python
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

```python
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

From the result above, the two items which is milk and bread has a 0.25 and 0.65 and a confidence of .80, which means there is an 80% chance that if a person bought milk, they will also buy bread.

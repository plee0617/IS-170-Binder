First Data Set - PreferredColor

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv('PreferredColor.csv')
df.head()
```

```python
# Missing a header so we have to create a new one, titled "Chosen Color"
df=pd.read_csv('PreferredColor.csv', names=['Chosen Color'])
df.head() 
```

```python
df.shape
```

```python
# Code will convert the dataframe into a list
data=list(df["Chosen Color"].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Any False or True will be read as 0 or 1 instead
df=df.replace(False,0)
df=df.replace(True,1)

df
```

```python
df=apriori(df,min_support=0.2,use_colnames=True)
df
```

```python
# Lift measures the correlation between antecedent and consequent when they are bought separately from each other. 
# Lift = 1 means that there is no correlation within the items.
# Lift > 1 means that there is a positive correlation within the items, i.e., products in the items,x and y, are more likely to be bought together.
# Lift < 1 means that there is a negative correlation within the items, i.e., products in items,x and y, are unlikely to be bought together.
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

```python
# The confidence is the probability of seeing the consequent (C) in same transaction with antecedent (A). 
# The confidence value 1 means antecedent and consequent always occur together (100%).
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

From the result shown above, the two items which are the color Brown and Blue has a 0.40 and 0.73 support and a confidence of .83, which means there is an 83% chance that if a person prefers the color Brown, they will also prefer the color Blue.
